```python
# Nama = Fakhrity Hikmawan
# NIM = 1103204074
# Kelas = TK-44-G4

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Membaca data dari file CSV
data = pd.read_csv('/content/sample_data/EconomiesOfScale.csv')

# Memisahkan variabel independen (x) dan variabel dependen (y)
x = data.iloc[:, 0:1].values
y = data.iloc[:, 1].values

# Membuat objek PolynomialFeatures dengan derajat polinomial yang diinginkan
degree = 2
poly_reg = PolynomialFeatures(degree=degree)

# Mengubah fitur input (x) menjadi fitur polinomial
x_poly = poly_reg.fit_transform(x)

# Membuat model regresi linear dengan fitur polinomial
model = LinearRegression()
model.fit(x_poly, y)

# Membuat prediksi berdasarkan model
y_pred = model.predict(x_poly)

# Plot hasil regresi polinomial
plt.scatter(x, y, color='blue', label='Data Asli')
plt.plot(x, y_pred, color='red', label='Regresi Polinomial')
plt.xlabel('Variabel Independen')
plt.ylabel('Variabel Dependan')
plt.title('Regresi Polinomial')
plt.legend()
plt.show()
```
